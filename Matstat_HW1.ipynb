# –ó–∞–¥–∞—á–∞ 1

### –ü—É–Ω–∫—Ç –∞)

```python
%matplotlib inline
import scipy.stats as st
import matplotlib.pyplot as plt
import collections


class MyDistribution(st.rv_continuous):

    def _cdf(self, x, **kwargs):
        if x <= 1:
            return 0
        else:
            return 1 - 1 / x


def printIterable(iterable):
    if not isinstance(iterable, collections.Iterable):
        raise TypeError("Argument should be iterable")
    iter = 0
    for i in iterable:
        iter += 1
        print(i, '\t', end='')
        if iter % 7 == 0:
            print()

my_disturb = MyDistribution()
```

–í—ã–±–æ—Ä–∫–∞ –æ–±—ä—ë–º–∞ 100:

```python
selection100 = my_disturb.rvs(size=100)
printIterable(selection100)
```

–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ –≤—ã–±–æ—Ä–∫–∏ –æ–±—ä—ë–º–∞ 100 —Å 10 —Å—Ç–æ–ª–±—Ü–∞–º–∏:

```python
plt.hist(selection100, bins=10)
plt.show()
```

–í—ã–±–æ—Ä–∫–∞ –æ–±—ä—ë–º–∞ 1000:

```python
selection1000 = my_disturb.rvs(size=1000)
printIterable(selection1000)
```

–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –≤—ã–±–æ—Ä–∫–∏ –æ–±—ä—ë–º–∞ 1000 —Å 10 —Å—Ç–æ–ª–±—Ü–∞–º–∏

```python
print("–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –≤—ã–±–æ—Ä–∫–∏ = ", max(selection1000))
plt.hist(selection1000, bins=10)
plt.show()
```

–ü–æ—á–µ–º—É –æ—Ç–ª–∏—á–∞—é—Ç—Å—è –≥—Ä–∞—Ñ–∏–∫–∏? (–æ—Ç–ª–∏—á–∏–µ - –≤–æ –≤—Å–µ—Ö —Å—Ç–æ–ª–±—Ü–∞—Ö, –∫—Ä–æ–º–µ –ø–µ—Ä–≤–æ–≥–æ, –∑–Ω–∞—á–µ–Ω–∏–π –ø–æ—á—Ç–∏ –Ω–µ—Ç)
–ü—Ä–∏ –±–æ–ª—å—à–µ–º –æ–±—ä—ë–º–µ –≤—ã–±–æ—Ä–∫–∏ –≤—ã—à–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å —Ç–æ–≥–æ, —á—Ç–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –≤—ã–±–æ—Ä–∫–∏ –±—É–¥–µ—Ç –æ—á–µ–Ω—å –±–æ–ª—å—à–æ–π. –ê —á–µ–º –±–æ–ª—å—à–µ —ç–ª–µ–º–µ–Ω—Ç –≤—ã–±–æ—Ä–∫–∏, —Ç–µ–º –º–µ–Ω—å—à–µ –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –µ–≥–æ –ø–æ—è–≤–ª–µ–Ω–∏—è, —Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ, –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –±—É–¥—É—Ç –¥–∞–ª–µ–∫–æ –æ—Ç —ç—Ç–æ–≥–æ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ —ç–ª–µ–º–µ–Ω—Ç–∞ (—Å–ª–µ–≤–∞ –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–µ). –ü–æ–ª—É—á–∞–µ—Ç—Å—è, —á—Ç–æ –±–æ–ª—å—à–∏–Ω—Å—Ç–≤–æ —ç–ª–µ–º–µ–Ω—Ç–æ–≤ –≤—ã–±–æ—Ä–∫–∏ –Ω–∞—Ö–æ–¥—è—Ç—Å—è –≤ –ª–µ–≤–æ–π —á–∞—Å—Ç–∏ –≥—Ä–∞—Ñ–∏–∫–∞ –≤ –ø–µ—Ä–≤–æ–º —Å—Ç–æ–ª–±—Ü–µ. –ê —á–µ–º –±–æ–ª—å—à–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç, —Ç–µ–º –±–æ–ª—å—à–∏–µ —ç–ª–µ–º–µ–Ω—Ç—ã –≤–∫–ª—é—á–µ–Ω—ã –∏–º–µ–Ω–Ω–æ –≤ –ø–µ—Ä–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü.

### –ü—É–Ω–∫—Ç –±)

–í—ã–±–æ—Ä–∫–∞ –∏–∑ 1000 —Ä–µ–∞–ª–∏–∑–∞—Ü–∏–π –°–í ùëå = $\sum_{i=1}^{30} X_i$, –≥–¥–µ $X_i$ ‚Äì –°–í, —Ä–∞—Å–ø—Ä–µ–¥–µ–ª—ë–Ω–Ω–∞—è –ø–æ –¥–∞–Ω–Ω–æ–º—É –≤ –∑–∞–¥–∞–Ω–∏–∏ –∑–∞–∫–æ–Ω—É:

```python
advanced_selection = []
for i in range(1000):
    advanced_selection.append(sum(my_disturb.rvs(size=30)))
```

```python
printIterable(advanced_selection)
```

–ì–∏—Å—Ç–æ–≥—Ä–∞–º–º–∞ —Ç–∞–∫–æ–π –≤—ã–±–æ—Ä–∫–∏:

```python
print("–ú–∞–∫—Å–∏–º–∞–ª—å–Ω—ã–π —ç–ª–µ–º–µ–Ω—Ç –≤—ã–±–æ—Ä–∫–∏ = ", max(advanced_selection))
plt.hist(advanced_selection, bins=10)
plt.show()
```
