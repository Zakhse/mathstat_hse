### Пункт а)

```python
%matplotlib inline
import scipy.stats as st
import matplotlib.pyplot as plt
import collections


class MyDistribution(st.rv_continuous):

    def _cdf(self, x, **kwargs):
        if x <= 1:
            return 0
        else:
            return 1 - 1 / x


def printIterable(iterable):
    if not isinstance(iterable, collections.Iterable):
        raise TypeError("Argument should be iterable")
    iter = 0
    for i in iterable:
        iter += 1
        print(i, '\t', end='')
        if iter % 7 == 0:
            print()

my_disturb = MyDistribution()
```

Выборка объёма 100:

```python
selection100 = my_disturb.rvs(size=100)
printIterable(selection100)
```

Гистограмма выборки объёма 100 с 10 столбцами:

```python
plt.hist(selection100, bins=10)
plt.show()
```

Выборка объёма 1000:

```python
selection1000 = my_disturb.rvs(size=1000)
printIterable(selection1000)
```

Гистограммы выборки объёма 1000 с 10 столбцами

```python
print("Максимальный элемент выборки = ", max(selection1000))
plt.hist(selection1000, bins=10)
plt.show()
```

Почему отличаются графики? (отличие - во всех столбцах, кроме первого, значений почти нет)
При большем объёме выборки выше вероятность того, что максимальный элемент выборки будет очень большой. А чем больше элемент выборки, тем меньше вероятность его появления, следовательно, большинство элементов будут далеко от этого максимального элемента (слева на графике). Получается, что большинство элементов выборки находятся в левой части графика в первом столбце. А чем больше максимальный элемент, тем большие элементы включены именно в первый столбец.

### Пункт б)

Выборка из 1000 реализаций СВ 𝑌 = $\sum_{i=1}^{30} X_i$, где $X_i$ – СВ, распределённая по данному в задании закону:

```python
advanced_selection = []
for i in range(1000):
    advanced_selection.append(sum(my_disturb.rvs(size=30)))
```

```python
printIterable(advanced_selection)
```

Гистограмма такой выборки:

```python
print("Максимальный элемент выборки = ", max(advanced_selection))
plt.hist(advanced_selection, bins=10)
plt.show()
```

Полученный график похож на второй график из пункта а).
